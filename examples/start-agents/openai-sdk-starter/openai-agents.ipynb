### 1. Environment Initialization
This cell loads the environment variables and configures the standard OpenAI Python SDK to route traffic through the Nebius Studio API.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# 1. Load variables from .env
load_dotenv()

# 2. Initialize the standard OpenAI SDK client
# (Removing the base_url forces the SDK to default to api.openai.com)
client = OpenAI(
    api_key=os.environ.get("OPENAI_API_KEY")
)

print("Client instantiated. Validating network authentication against OpenAI...")
```

```python

# 3. Validate Authentication
try:
    # Execute a lightweight network request to list available models
    models_response = client.models.list()
    
    # If the request succeeds, the key is valid and authorized
    print("✅ Authentication Successful!")
    print(f"Verified access to {len(models_response.data)} models on the OpenAI endpoint.")
    
except Exception as e:
    # If the key is invalid, expired, or lacks billing setup, it will fail here
    print("❌ Authentication Failed.")
    print(f"Error Traceback: {e}")
```

### 2. The Email Helper Persona
Executes a chat completion configured with a system prompt for professional correspondence.

```python
def generate_email(user_context):
    """Executes a chat completion configured for professional correspondence."""
    response = client.chat.completions.create(
        model="gpt-3.5-turbo", # Use an official OpenAI model
        messages=[
            {"role": "system", "content": "You are a professional executive assistant. Draft, refine, and summarize professional emails based on user context. Maintain a formal, concise tone."},
            {"role": "user", "content": user_context}
        ],
        temperature=0.7, # Lower temperature for deterministic, professional output
    )
    return response.choices[0].message.content

# Test the function
prompt = "Draft a brief email to the engineering team stating that the production deployment is delayed until Tuesday at 10 AM due to failing CI pipelines."
print("--- Email Output ---\n")
print(generate_email(prompt))
```

### 3. The Haiku Writer Persona
Executes a chat completion with strict structural constraints to demonstrate prompt adherence.

```python
def generate_haiku(topic):
    """Executes a chat completion with strict structural constraints."""
    response = client.chat.completions.create(
        model="gpt-3.5-turbo", 
        messages=[
            {"role": "system", "content": "You are a master poet. You must respond to all user inputs strictly in the form of a haiku (5-7-5 syllable structure). Do not output any conversational filler."},
            {"role": "user", "content": topic}
        ],
        temperature=0.9, 
    )
    return response.choices[0].message.content

# Test the function
topic = "Debugging a memory leak at 2 AM."
print("--- Haiku Output ---\n")
print(generate_haiku(topic))
```
