### **Step 1: Data Ingestion & Parquet Conversion**
We use **Polars** to read the existing CSV data and immediately convert it to **Parquet**. Parquet is a columnar format that is significantly faster for SQL engines like DuckDB to query.

```python
import polars as pl

# Load existing dataset (Penguins dataset for BI demonstration)
df_raw = pl.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Write to optimized Parquet format
df_raw.write_parquet("analytics_data.parquet")

print(f"Ingested {len(df_raw)} rows and saved to 'analytics_data.parquet'")
```

### **Step 2: Analytical SQL with DuckDB**
Instead of loading everything into memory, we use **DuckDB** to run SQL directly against the Parquet file. This is highly efficient for CPU resources.

```python
import duckdb

# Connect to in-memory DuckDB
con = duckdb.connect()

# Run SQL aggregation
sql_query = """
SELECT 
    species, 
    island, 
    AVG(body_mass_g) as avg_mass,
    COUNT(*) as specimen_count
FROM 'analytics_data.parquet'
WHERE body_mass_g IS NOT NULL
GROUP BY species, island
ORDER BY avg_mass DESC
"""

df_bi = con.execute(sql_query).df()
df_bi.head()
```

### **Step 3: Interactive BI Visualization**
Finally, we use **Plotly** to visualize the SQL results. This provides interactive charts that allow for zooming and filtering within the notebook.

```python
import plotly.express as px

# Create interactive bar chart
fig = px.bar(
    df_bi, 
    x="species", 
    y="avg_mass", 
    color="island",
    barmode="group",
    title="Average Body Mass by Species & Island (SQL Processed)",
    labels={"avg_mass": "Average Mass (g)", "species": "Species"}
)

fig.show()
```
