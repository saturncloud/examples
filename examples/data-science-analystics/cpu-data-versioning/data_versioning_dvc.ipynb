### **Step 1: Initialize DVC**
Set up the local environment to track data metadata.

```python
!dvc init --no-scm -f
print("DVC initialized successfully.")
```

### **Step 2: Configure Remotes (Local vs. Remote S3)**
We use a local folder to simulate S3 for testing, but provide the logic for a real S3 bucket.

```python
import os

# --- OPTION A: LOCAL SIMULATION (Free) ---
local_storage = "/tmp/dvc_simulated_s3"
os.makedirs(local_storage, exist_ok=True)
!dvc remote add -d local_remote {local_storage} -f

# --- OPTION B: REMOTE S3 BUCKET (Production Setup) ---
# To use a real bucket, uncomment the lines below and provide your URI
# S3_URI = "s3://your-real-bucket-name/data-folder"
# !dvc remote add -d s3_remote {S3_URI} -f

print(f"✅ Active Remote set to: {local_storage}")
```

### **Step 3: Track, Version, and Push Data**
We create a dummy dataset and 'push' it to our remote simulation.

```python
import pandas as pd

# 1. Create dataset
df = pd.DataFrame({'feature_x': [10, 20, 30], 'label': [1, 0, 1]})
df.to_csv('my_dataset.csv', index=False)

# 2. Track with DVC
!dvc add my_dataset.csv

# 3. Push to storage (Local Sim or S3)
!dvc push

print("\nData pushed to remote storage. Tracking file 'my_dataset.csv.dvc' created.")
```

### **Step 4: The Recovery Test**
Prove reproducibility by deleting the local data and pulling it back from the remote.

```python
os.remove('my_dataset.csv')
print("Local dataset deleted.")

!dvc pull
print("\n✅ Data successfully recovered from remote storage!")
```
