### **Step 2: Load and Prepare the Dataset**
We load the **Air Passengers** CSV file directly from a verified online repository. We use Pandas to parse the 'Month' column as a date index, which is essential for accurate time-series plotting.

```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np

# Load data and set the Month as the index
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df = pd.read_csv(url, header=0, index_col=0, parse_dates=[0])
df.columns = ['Passengers']
df.head()
```

### **Step 3: Generate a Simple Forecast Trend**
To visualize the underlying growth of the data, we calculate a **12-month Rolling Mean**. This "toy forecast" smooths out monthly seasonal noise to show the general direction of passenger growth over the years.

```python
# Calculate rolling average to reveal the trend
df['Trend_Forecast'] = df['Passengers'].rolling(window=12).mean()
```

### **Step 4: Build the Interactive Visualization**
Finally, we use Plotly to create an interactive chart. We include a **Range Slider** at the bottom for quick zooming and **Range Selectors** (1y, 5y) to instantly jump to specific time windows. Hovering over the lines will show exact values for both actual data and the trend forecast.

```python
# Initialize figure with actual and forecast traces
fig = go.Figure()

fig.add_trace(go.Scatter(x=df.index, y=df['Passengers'], name='Actual Passengers', mode='lines'))
fig.add_trace(go.Scatter(x=df.index, y=df['Trend_Forecast'], name='12m Forecast', line=dict(dash='dot')))

# Configure Layout for Interactivity
fig.update_layout(
    title='Interactive Air Passengers: Zoom + Forecast',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1y", step="year", stepmode="backward"),
                dict(count=5, label="5y", step="year", stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(visible=True), 
        type="date"
    ),
    hovermode='x unified',
    template='plotly_white'
)

fig.show()
```
